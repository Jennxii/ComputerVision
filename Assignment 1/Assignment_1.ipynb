## Halving

Load a colored image and halve each color value. Save the image and compare it with the original image.

```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('bird_1.jpg')

#halving color values
halved_image = image // 2

halved_image = cv2.cvtColor(halved_image, cv2.COLOR_BGR2RGB)

cv2.imwrite('halved_bird.jpg', halved_image)
plt.imshow(halved_image)
#plt.axis('off')  # Turn off axis numbers and ticks
#plt.show()
```

## Invert 

Load a colored image and invert the color values. 
Invert: $a_{max} - a$

```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('clinton_1.jpg')

b, g, r = cv2.split(image)

# Invert the color values for each channel
b = 255 - b
g = 255 - g
r = 255 - r

# Merge the inverted channels
inverted = cv2.merge((b, g, r))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


cv2.imwrite('clinton_1.jpg', image)
plt.imshow(image)
```

# Color

Load a colored image and set all the values of the channels to 0 except of one.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('trump_1.jpg')

# Set all channels to 0 except for the green channel
blue_channel = np.zeros_like(image[:, :, 0])
red_channel = np.zeros_like(image[:, :, 0])

# Create the mask for the green channel
green_mask = np.ones_like(image[:, :, 0])
blue_mask = np.zeros_like(image[:, :, 0])
red_mask = np.zeros_like(image[:, :, 0])

# Merge the channels with the masks
result_image = cv2.merge((blue_channel, green_mask * image[:, :, 1], red_channel))

result_image = cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB)


cv2.imwrite('trump_1.jpg', result_image)
plt.imshow(result_image)
```

## Flip

Load a colored image and flip the image along the y-axis.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('cat.png')

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image = cv2.flip(image,1)
cv2.imwrite('cat.png', image)
plt.imshow(image)
```

## Diff Image

Load two images with the same resolution and compute the means squared error (MSE) of them. Calculate the MSE for every color channel separately. For each color channel, calculate the difference of the pixels at the same position. The difference is squared. The squared differences are summed up. The sum is divided by the number of summed values.



```python
import cv2
import numpy as np

# Load the two images
image1 = cv2.imread('obama_2.jpg')
image2 = cv2.imread('obama_3.jpg')

# Compute the MSE for each color channel
mse_b = np.mean((image1[:,:,0] - image2[:,:,0])**2)
mse_g = np.mean((image1[:,:,1] - image2[:,:,1])**2)
mse_r = np.mean((image1[:,:,2] - image2[:,:,2])**2)

# Compute the overall MSE
overall_mse = np.mean((image1 - image2)**2)

# Print the MSE for each color channel and the overall MSE
print(f"MSE for Blue channel: {mse_b}")
print(f"MSE for Green channel: {mse_g}")
print(f"MSE for Red channel: {mse_r}")
print(f"Overall MSE: {overall_mse}")
```
